```python
#Because the data is a time series, the nearest value is used instead of the np.nan value
fill_list = ['Inflation_rate','Instsred_rate','YounthUnempRate','Gold_price']

for name in fill_list:
    temp = None
    index = train_data.shape[0]
    for ele in reversed(train_data[name]):
        #print(ele)
        index -= 1
        if pd.notnull(ele) == True:
            temp = ele
            break
    #print(index , temp)
    train_data[name] = train_data[name].fillna(temp)
    
train_data

    
```

```python
#delete listTime and separate label&feature  (function : iloc 和 slice 操作)
all_feature = train_data.iloc[:,2:]
all_label = train_data['GDP_China']

```

```python


numeic_feature = all_feature.dtypes[all_feature.dtypes != 'object'].index

numeic_feature
```

```python
#DATA standardization(Z-zero)
all_feature[numeic_feature] = all_feature[numeic_feature].apply(
                    lambda x : (x - x.mean()) / x.std() 
)
```

```python
#creat Tensor
train_feature = torch.tensor(all_feature.values,dtype=torch.float32)
train_label = torch.tensor(all_label.values,dtype=torch.float32)

#  float32 !!!很重要,DL常用32
```

```python
#craet Tensordataset + compute iteration
def load_array(dataarray,batach_size,is_train=True): #dataarray include feature & labels
    dataset = data.TensorDataset(*dataarray) # paramter : Tuple[Tensor,...]
    return data.DataLoader(dataset,batach_size,shuffle=is_train)
```

```python
Loss = nn.MSELoss()
#Single layer linear perceptron
def get_net():
    net = nn.Sequential(nn.Linear(train_feature.shape[1],1))
    return net

def log_rmse(net,feature,label):
    clipps_preds = torch.clamp(net(feature),1,float('inf'))
    rmse = torch.sqrt(Loss(torch.log(clipps_preds),torch.log(label)))
    return rmse
```

```python
def train(net,train_feature,train_label,test_feature,test_label,num_epoch,learning_rate,weight_decay,batach_size):
    #test_feature,test_label 是为了方便交叉验证
    train_loss,test_loss=[],[]
    train_iteration = load_array((train_feature,train_label),batch_size)
    optimizer = torch.optim.Adam(net.parameters(),lr=learning_rate,weight_decay=weight_decay)
    for epoch in range(num_epoch):
        for X,y in train_iteration:
            optimizer.zero_grad()
            l=Loss(net(X),y)
            l.backward()
            optimizer.step()
        train_loss.append(log_rmse(net,train_feature,train_label))
        if test_label is not None:
            test_loss.append(log_rmse(net, test_feature, test_label))
    return train_loss,test_loss
```

```python
def sliceTrainValid_k_fold(k,i,X,y):
    assert k >1
    X_train,y_train = None,None
    fold_size = X.shape[0] // k
    for j in range(k):
        sli=slice(j * fold_size,(j+1) * fold_size)
        X_part = X[sli,:]
        y_part = y[sli]
        if j==i:
            X_valid = X_part
            y_vaild = y_part
        elif X_train is None:
            X_train = X_part
            y_train = y_part
        else:
            X_train = torch.cat([X_train,X_part],0)
            y_train = torch.cat([y_train,y_part],0)
    return X_train,y_train,X_valid,y_vaild
            
```

```python
def k_times_k_fold_CrossValidation(k,X_train,y_train,num_epoch,learning_rate,weight_decay,batch_size):
    train_loss_sum , valid_loss_sum = 0,0
    for i in range(k):
        data = sliceTrainValid_k_fold(k,i,X_train,y_train)
        net = get_net()
        train_loss,valid_loss = train(net, *data,num_epoch,learning_rate,weight_decay,batch_size)
        train_loss_sum += train_loss[-1]
        valid_loss_sum += valid_loss[-1]
        print(f'fold-{i+1} trainset log_rmse{float(train_loss[-1]):f}  '+ f'fold-{i+1} validset log_rmse{float(valid_loss[-1]):f}')
    return train_loss_sum/k , valid_loss_sum/k
              
```

```python
#Debug hyperparameter

k , num_epoch,learning_rate,weight_decay,batch_size=6,100,5,0,1

train_averg_loss,valid_averg_loss =k_times_k_fold_CrossValidation(k ,train_feature,train_label,num_epoch,learning_rate,weight_decay,batch_size)
print(f'train_averg_loss{float(train_averg_loss):f},  ' +f'valid_averg_loss{float(valid_averg_loss):f}')
```

```python
def trainInAll_and_pred(train_feature, test_feature, train_label,test_data,num_epochs, lr, weight_decay, batch_size):
    net = get_net()
    train_loss,_ = train(train_feature,train_label,None,None,num_epochs, lr, weight_decay, batch_size)
    print(f'train_lossMse {float(train_loss[-1]):f}  ')
    predicted = net(test_feature).detach().numpy()
    test_data['ChinaGDP_predicted']=pd.Series(predicted.reshape(1,-1)[0])
    test_data.to_csv('result.csv',index=False)
```
